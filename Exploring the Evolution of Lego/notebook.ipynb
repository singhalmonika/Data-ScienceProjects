```python
# Use this cell to begin your analyses, and add as many cells as you would like!
import pandas as pd
lego_set= pd.read_csv("datasets/lego_sets.csv")
lego_set.head()
```

```python
#Inspect the dataset
lego_set.shape
lego_set.info()
lego_set.duplicated()
```

```python
# Drop all the duplicate values from the dataframe
lego_set = lego_set.drop_duplicates()
lego_set.head()
lego_set.shape
```

```python
# Drop NA value from set number column
lego_set=lego_set.dropna(subset=['set_num'])
lego_set.head()
lego_set.shape
lego_set.info()
```

```python
# Import parent_theme csv file
parent_theme_set = pd.read_csv("datasets/parent_themes.csv")
parent_theme_set.head()
```

```python
#check the info
parent_theme_set.shape
parent_theme_set.info()
```

```python
# join the two datasets left:parent_theme and right_name
lego_joined_dataset=lego_set.merge(parent_theme_set, left_on='parent_theme', right_on='name')
lego_joined_dataset.head()
```

```python
#Print the shape of the joined dataset
lego_joined_dataset.shape
lego_joined_dataset.info()
lego_joined_dataset.isna().sum()
```

```python
lego_licensed_join_dataset=lego_joined_dataset[lego_joined_dataset["is_licensed"]]
lego_licensed_join_dataset.head()
```

```python
grouped_legoDataframe = (
    lego_licensed_join_dataset
    .groupby(by=['id','parent_theme'])
    .count()
    ["set_num"]
    .reset_index()
     )
grouped_legoDataframe
```

```python
#grouped_legoDataframe[grouped_legoDataframe['parent_theme']=="Star Wars"]
total = grouped_legoDataframe.set_num.sum()
print(total)
grouped_legoDataframe["perc"] = grouped_legoDataframe.set_num*100/total
grouped_legoDataframe
```

```python
the_force = (grouped_legoDataframe
             .query("parent_theme=='Star Wars'")
             .perc[0]
            )
the_force
```

```python
gp_year_setnum=(
    lego_licensed_join_dataset
    .groupby(by=['year','id','parent_theme'])
    .count()
    ["set_num"]
    .reset_index()
     )
gp_year_setnum
```

```python
pv_table = pd.pivot_table(gp_year_setnum,values='set_num', index=['year'],
                    columns=['parent_theme'])
pv_table
```

```python
starwr_series = pv_table.loc[:,'Star Wars']
```

```python
max_ser= pv_table.max(axis=1)
```

```python
new_era = int(pv_table[starwr_series<max_ser].
              reset_index().
              loc[:,'year'][0])
new_era
```
